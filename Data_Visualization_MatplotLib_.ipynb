# Histogram  
A Histogram is the most commonly used graph to show frequency distributions and a frequency distribution shows how often each different value in a set of data occurs.

The histogram represents the number of times the value between a particular range has occurred and are plotted as one bar for that range.

# Stack Plot or stack bar chart
To understand stacked bar chart easily, I had used a very interesting example from our school days. Do you remember sometimes we use to get grace marks (not to pass but) because of one or more questions were wrong or out of syllabus. I'll be using stacked bar chart to display those Grace marks stacked on top of the actual marks obtained by the students. Further, we can apply various properties such as change the colour of stacked potion, or the actual base bar of plot in matplotlib

````python
import matplotlib.pyplot as plt
avg_scr=[55,59,67,29,36,98,45,78,88,80]
sub=["Math", "Phy", "Che", "Bio", "IT", "Sci", "Tec", "PT", "ECE", "CSE"]
grace_mark=[15,12,10,25,15,-25,12,-5,-2, -5]

plt.title("Average Marks for each subject", color="r", fontsize=14)
plt.xlabel("Subjects", color="r", fontsize=14)
plt.ylabel("Average Marks", color="m", fontsize=14)

plt.bar(sub, avg_scr, width = 0.7, 
        color="Blue",
        label="Average Score")
plt.bar(sub, grace_mark, bottom = avg_scr, 
        color="Red",
        label="Grace Mark")

plt.legend(loc="best")
plt.show()
````

````python
import matplotlib.pyplot as plt
avg_scr=[55,59,67,29,36,98,45,78,88,80]
sub=["Math", "Phy", "Che", "Bio", "IT", "Sci", "Tec", "PT", "ECE", "CSE"]
grace_mark=[15,12,10,25,15,-25,12,-5,-2, -5]

plt.title("Average Marks for each subject", color="r", fontsize=14)
plt.xlabel("Subjects", color="r", fontsize=14)
plt.ylabel("Average Marks", color="m", fontsize=14)

plt.bar(sub, avg_scr, width = 0.7, 
        color="Blue",
        label="Average Score",
        yerr=grace_mark)
plt.bar(sub, grace_mark, bottom = avg_scr, 
        color="Red",
        label="Grace Mark")

plt.legend(loc="best")
plt.show()
````

# 3D data Visualization 

Beside able to project the linear and nonlinear data in 2D plot, Matplotlib is also capable of visualising the data in three-dimensional plot. Matplotlib can handle 3-Dimensional data efficiently, so we need XY and Z axis to plot 3D data. It is well known fact that human brain perceives 3D data better than any other form of data for visualisation. It's because we live in 3D world where height, width and depth are the part of visual dimension. In this lesson you will learn to create sample data which can be displayed in XY plot and then combining all these data into a three-dimensional space.
````python
from dash import Dash, dcc, html, Input, Output
import plotly.express as px

app = Dash(__name__)


app.layout = html.Div([
    html.H4('Interactive scatter plot with Iris dataset'),
    dcc.Graph(id="scatter-plot"),
    html.P("Filter by petal width:"),
    dcc.RangeSlider(
        id='range-slider',
        min=0, max=2.5, step=0.1,
        marks={0: '0', 2.5: '2.5'},
        value=[0.5, 2]
    ),
])


@app.callback(
    Output("scatter-plot", "figure"), 
    Input("range-slider", "value"))
def update_bar_chart(slider_range):
    df = px.data.iris() # replace with your own data source
    low, high = slider_range
    mask = (df['petal_width'] > low) & (df['petal_width'] < high)
    fig = px.scatter(
        df[mask], x="sepal_width", y="sepal_length", 
        color="species", size='petal_length', 
        hover_data=['petal_width'])
    return fig


app.run_server(debug=True)
````
